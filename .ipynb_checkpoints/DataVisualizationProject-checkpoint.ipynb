## Question 2

```python
# sort the data in ascending order
df_survey.sort_values(['Very interested'],ascending=False,axis=0,inplace=True)
df_survey
```

```python
# convert the numbert to percentage of total
df_survey_percentage=df_survey/2233
df_survey_percentage
```

```python
# round to two decimal point
df_survey_PercentRound=df_survey_percentage.round(2)
df_survey_PercentRound
```

```python
ax = df_survey_PercentRound.plot(kind='bar',
                                 width=0.8,
                                 figsize=(20, 8),
                                color=['#5cb85c','#5bc0de', '#d9534f'], fontsize=14)

ax.set_title('Percentage of Responders\' Interest in Data Science Areas', fontsize=16)
ax.legend(fontsize=14)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.axes.get_yaxis().set_visible(False)
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate('{:.0%}'.format(height), (x+width/4, y + height+0.01))
```
